```python
# Step 4: Read the research paper text from a .docx file
from docx import Document

def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for paragraph in doc.paragraphs:
        full_text.append(paragraph.text)
    return "\n".join(full_text)
```

```python
# Replace 'MyDrive/path_to_your_file/research_paper.docx' with your file path
file_path = "/content/drive/MyDrive/Final thesis_Shivani.docx"
```

```python
# Read the research paper text
research_paper_text = read_docx(file_path)
```

```python
# Step 6: Summarize the text
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# Step 6: Summarize the text
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
# %%
MAX_INPUT_LENGTH = 1000  # Adjust this based on the model's token limit # Changed AX_INPUT_LENGTH to MAX_INPUT_LENGTH
chunks = [research_paper_text[i:i+MAX_INPUT_LENGTH] for i in range(0, len(research_paper_text), MAX_INPUT_LENGTH)]

summaries = []
for chunk in chunks:
    summary = summarizer(chunk, max_length=50, min_length=10, do_sample=False)[0]['summary_text']
    summaries.append(summary)
```

```python
# Combine the summaries
final_summary = " ".join(summaries)

# Step 7: Print the summary
print(final_summary)
```
