```python
# Step 2: Import necessary modules
from sentence_transformers import SentenceTransformer, util
```

```python
# Step 3: Read the research paper text from a .docx file
from docx import Document

def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for paragraph in doc.paragraphs:
        full_text.append(paragraph.text)
    return "\n".join(full_text)
```

```python
# Path to your reference paper in Google Drive
reference_file_path = "/content/drive/MyDrive/Literature.docx"
reference_text = read_docx(reference_file_path)
```

```python
# Step 4: Read your thesis document (same method)
thesis_file_path = "/content/drive/MyDrive/Final thesis_Shivani .docx"
thesis_text = read_docx(thesis_file_path)
```

```python
# Step 5: Split the thesis and reference text into chunks
thesis_chunks = thesis_text.split("\n\n")  # Split thesis into paragraphs or smaller sections
reference_chunks = reference_text.split("\n\n")  # Split reference paper similarly
```

```python
# Step 6: Load Sentence Transformer model for semantic similarity
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# Step 7: Compare Thesis Chunks with Reference Paper
similarity_scores = []
for chunk in thesis_chunks:
    chunk_embedding = model.encode(chunk, convert_to_tensor=True)
    reference_embeddings = model.encode(reference_chunks, convert_to_tensor=True)

    # Compute cosine similarity and take the maximum score
    similarities = util.pytorch_cos_sim(chunk_embedding, reference_embeddings)
    max_similarity = similarities.max().item()  # Maximum similarity for this chunk
    similarity_scores.append(max_similarity)
```

```python
# Step 8: Calculate overall plagiarism score
overall_score = sum(similarity_scores) / len(similarity_scores)
print(f"Plagiarism Score: {overall_score * 100:.2f}%")
```

```python
# Step 9: Interpret the plagiarism score
if overall_score > 0.7:
    print("High plagiarism detected!")
elif overall_score > 0.4:
    print("Moderate plagiarism detected.")
else:
    print("Low plagiarism detected.")
```
