```python
import pandas as pd
```

```python
movies=pd.read_csv('/root/.cache/kagglehub/datasets/ahsanaseer/top-rated-tmdb-movies-10k/versions/1/top10K-TMDB-movies.csv')
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'overview', 'genre']]
```

```python
movies.head()
```

```python
movies['class']=movies['overview']+movies['genre']
```

```python
movies
```

```python
new_movie_data  = movies.drop(columns=['overview', 'genre'])
```

```python
new_movie_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer       #tokenization,vocablury building and frequency count
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector=cv.fit_transform(new_movie_data['class'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_movie_data[new_movie_data['title']=="The Last Airbender"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_movie_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    # Calculate similarity scores for the movie with all other movies
    # 'similarity' is assumed to be a precomputed matrix of similarity scores
    # Enumerate pairs each movie's index with its similarity score
    # Sort the list of tuples in descending order of similarity scores
    index=new_movie_data[new_movie_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_movie_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
