```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
#Load model and Tokenizer
model=AutoModelForCausalLM.from_pretrained("microsoft/Phi-3-mini-4k-instruct",
                                           device_map='cuda',         #Specifies the device to use for computation. 'cuda' ensures the model is loaded on the GPU for faster processing.
                                           torch_dtype='auto',        #	Automatically sets the data type for tensors (e.g., float16 if supported, otherwise float32). Useful for optimizing performance
                                           trust_remote_code=True,    #Allows loading custom Python code from the model's repository (needed if the repository contains unique implementations).
                                           )
tokenizer=AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
from transformers import pipeline  # A utility from Hugging Face that simplifies using models for specific tasks like text generation

generator=pipeline("text-generation", #Specifies the task type as "text generation"
                    model=model,      #The pre-loaded causal language model to be used for generating text.
                    tokenizer=tokenizer,   # The tokenizer corresponding to the model for encoding/decoding text.
                    return_full_text=False, # If False, the generated text does not include the input prompt (only new text is returned).
                    max_new_tokens=500,    # Limits the maximum number of tokens in the generated output (ensures the response length is controlled).
                    do_sample=False        # If False, uses greedy decoding (deterministic); if True, enables stochastic decoding (e.g., sampling-based).
)
```

```python
messages = [  # Define a list of messages to simulate a conversation or input-output structure
    {  # Start of a dictionary representing a single message
        "role": "user",               # Specifies the role of the entity sending the message ("user" in this case).
        "content": "Create Funny Joke about friends."  # The content of the user's message (input for the system).
    }
]
```

```python
output=generator(messages)
print(output[0]["generated_text"])
```
